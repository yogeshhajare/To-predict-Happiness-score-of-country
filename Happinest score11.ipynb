### since the model is continued data... we use linear regression model

```python
# Linear Regression
from sklearn.linear_model import LinearRegression
lr=LinearRegression()

#Train test split model selection 
from sklearn.model_selection import train_test_split

# Calling Metrics 
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
```

```python
for i in range(1,100):
    x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=i)
    lr.fit(x_train,y_train)
    pred_train=lr.predict(x_train)
    pred_test=lr.predict(x_test)
    print("At random State",i)
    print("At random state, the Train accuracy is : ",round(r2_score(y_train,pred_train)*100,2))
    print("At random state, the Test accuracy is : ",round(r2_score(y_test,pred_test)*100,2))
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=32)
lr.fit(x_train,y_train)
```

```python
pred_train=lr.predict(x_train)
pred_test=lr.predict(x_test)
```

```python
print("R2 score",r2_score(y_test,pred_test)*100)
```

# Cross_Validation Model

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.svm import SVR
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import AdaBoostRegressor
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.model_selection import cross_val_score

dtc=DecisionTreeRegressor()
knn=KNeighborsRegressor()
svr=SVR()
gnb=GaussianNB()
ada=AdaBoostRegressor()
gbr=GradientBoostingRegressor()

model=[lr,dtc,knn,svr,gnb,ada,gbr]
```

```python
train_acc=r2_score(y_train,pred_train)
test_acc= r2_score(y_test,pred_test)

for i in range(2,10):
    cv=cross_val_score(lr,x,y,cv=i)
    cv_mean=cv.mean()
    print(f'Cross Fold {i} The cv score is {cv_mean} Training Accuracy is {train_acc} Test Accuracy is {test_acc}')
    print("\n")
```

# Ensemble Technique
